### 일변수 벡터함수
 $$r(t) = (f_1(t), f_2(t),···,f_n(t))^T$$

```python
# 3차원 벡터함수 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d  # 3차원 그래프를 그리기 위한 툴킷
import matplotlib as mpl

mpl.style.use('bmh')         # 그래프에 사용할 테마 설정
mpl.style.use('seaborn-whitegrid')


plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')    # 3차원으로 투영되는 좌표 생성
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)

# 입력값 t의 범위 설정
t = np.linspace(0,2,101)
X = np.sin(6*t)
y = 1/4*t
z = t**2/2

ax.plot3D(X,y,z,c='k')
ax.plot([X[0]], [y[0]], [z[0]], 'o', markersize=10, color='k',
        label=f't={t[0]:.2f}')
ax.plot([X[50]], [y[50]], [z[50]], '^', markersize=10, color='k',
        label=f't={t[50]:.2f}')
ax.plot([X[-1]], [y[-1]], [z[-1]], '*', markersize=10, color='k',
        label=f't={t[-1]:.2f}')

ax.legend(fontsize=15, loc='upper left')

plt.show()
```

### 다변수 스칼라함수
$$y = f(X)$$
예시 : $$f(x,y) = x+y$$

```python
# 다변수 스칼라함수 그리기
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d  # 3차원 그래프를 그리기 위한 툴킷
import matplotlib as mpl

mpl.style.use('bmh')         # 그래프에 사용할 테마 설정
mpl.style.use('seaborn-whitegrid')

plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel(r'$x_1$', fontsize=20)
ax.set_ylabel(r'$x_2$', fontsize=20)
ax.set_zlabel(r'$z$', fontsize=20)

# 함수의 정의역 설정
X1 = np.linspace(-2,2,50)
X2 = np.linspace(-1,3,50)
X1, X2 = np.meshgrid(X1, X2)
Z = 50*(X2-X1**2)**2 + (2-X1)**2

# 계산된 값을 3차원 공간에 점으로 나타내기
ax.scatter3D(X1, X2, Z, marker='.', color='gray')

plt.show()
```
